```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filename='习题一.xlsx'
data=pd.read_excel(filename)
data
```

```python
age=data.age
oxy=data.OXY
time=data.time
```

## 第一小题

```python
plt.scatter(oxy,age)
plt.xlabel('oxy')
plt.ylabel('age')
plt.title('oxy with age')
```

```python
plt.scatter(oxy,time)
plt.xlabel('oxy')
plt.ylabel('time')
plt.title('oxy with time')
```

可以看出肺活量与年龄没有必然关系，但是肺活量与跑1.5英里时间成一定的反比例关系。

## 第二小题

```python
pd.plotting.scatter_matrix(data,figsize=(12,12))
```

可以看出脉搏和最大脉搏呈现正相关，肺活量与跑步用时负相关，别的参数基本没有什么线性关系。
